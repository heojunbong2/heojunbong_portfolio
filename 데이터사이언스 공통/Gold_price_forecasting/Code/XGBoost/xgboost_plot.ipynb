```python
import matplotlib.pyplot as plt 
plt.title('Real vs Prediction',fontsize=15)
plt.plot(np.array(data['Date']),np.array(data['xgboost_feature_engineering_prediction']),label='xgboost_feature_engineering_prediction')
plt.plot(np.array(data['Date']),np.array(data['xgboost_original_prediction']),label='xgboost_original_prediction')
plt.plot(np.array(data['Date']),np.array(data['Close']),label='Real')
plt.xticks(rotation=45)
plt.legend(loc='upper center', bbox_to_anchor=(1.4, 1), ncol=1)  # 범례를 그래프 아래로 이동
plt.show()
```
