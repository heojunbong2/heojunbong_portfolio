```python
whole_data=pd.concat([df_2021,df_2022,df_2023])
whole_data

```

```python
g = nx.karate_club_graph()
g = nx.from_pandas_edgelist(whole_data, source = 'word', target = 'year')
pos = nx.spring_layout(g, k = 0.1, iterations=200)

color_map_node = []
for node in g:
  if node == 2021:
    color_map_node.append('yellow')
  elif node == 2022:
    color_map_node.append('green')
  elif node == 2023:
    color_map_node.append('orange')
  else:
    color_map_node.append('white')

color_map_edge = []
for edges in g.edges():
    if 2021 in edges:
        color_map_edge.append('yellow')
    if 2022 in edges:
        color_map_edge.append('green')
    if 2023 in edges:
        color_map_edge.append('orange')

plt.figure(figsize=(30, 30))

nx.draw_networkx_nodes(g, pos, node_size = 5000, node_color = color_map_node)
nx.draw_networkx_edges(g, pos, width=1, edge_color = color_map_edge)
nx.draw_networkx_labels(g, pos,  font_size=30, font_family='Malgun Gothic')

plt.show()
```

```python
import networkx as nx
import pandas as pd

df= pd.DataFrame(whole_data)


# 그래프 생성

g= nx.from_pandas_edgelist(df, source='word', target='year')

# Betweenness Centrality 계산
betweenness_centrality = nx.betweenness_centrality(g)

# Betweenness Centrality를 데이터프레임으로 변환
df_centrality = pd.DataFrame(list(betweenness_centrality.items()), columns=['word', 'betweenness_centrality'])

# 결과 출력
print(df_centrality)
```
