```python
import pandas as pd
from datetime import datetime


# Define a function to convert Korean month names to numeric format
def convert_month(korean_month):
    # Convert Korean month to English month
    english_month = datetime.strptime(korean_month, '%Y년 %m월 %d일').strftime('%Y-%m')
    return english_month

# Apply the function to the 'month' column
df['month'] = df['날짜'].apply(convert_month)

# Print the modified DataFrame
df.head(5)
```

```python
import pandas as pd


# Convert the 'date' column to a datetime object
df['month'] = pd.to_datetime(df['month'])

# Define bins for quarters
bins = pd.to_datetime(['2021-08-01', '2021-11-01', '2022-02-01', '2022-05-01', '2022-08-01',
                       '2022-11-01', '2023-02-01', '2023-05-01', '2023-08-01'])

# Create a new column 'quarters' based on the specified criteria
df['quarters'] = pd.cut(pd.to_datetime(df['month']), bins=bins, labels=['2021-08~2021-10', '2021-11~2022-01', '2022-02~2022-04', '2022-05~2022-07',
                                                                         '2022-08~2022-10', '2022-11~2023-01', '2023-02~04', '2023-05~2023-07'], right=False)

df
```

```python
from gensim.models import CoherenceModel
import matplotlib.pyplot as plt 

def show_coherence(corpus, dictionary, start=6, end=15):
    iter_num=[]
    per_value=[]
    coh_value=[]
    
    for i in range(start, end+1):
        model=LdaModel(corpus=corpus, id2word=dictionary, num_topics=i)
        iter_num.append(i)
        pv=model.log_perplexity(corpus)
        per_value.append(pv)
        
        cm=CoherenceModel(model=model, corpus=corpus, coherence='u_mass')
        cv=cm.get_coherence()
        coh_value.append(cv)
        print(f'num_topics: {i}, perplexity: {pv:0.3f}, coherence: {cv:0.3f}')
    
    plt.plot(iter_num, per_value, 'g-')
    plt.xlabel("num_topics")
    plt.ylabel("perplexity")
    plt.show()
    
    plt.plot(iter_num, coh_value, 'r-')
    plt.xlabel("num_topics")
    plt.ylabel("coherence")
    plt.show()
    
```

```python
show_coherence(corpus, dictionary, start=6,end=15)

# 최적의 토픽의 수 결정하기 coherence가 높은 지점인 3로 토픽 수 정하기
```
